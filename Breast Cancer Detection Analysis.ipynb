```python
### EXPLORATORY DATA ANALYSIS
# first exploring the output column
sns.countplot(df['diagnosis'])
```

```python
df_temp = df.drop(columns=['diagnosis'],axis=1) #dropping categorial attribute so df_temp only contains numerical attributes
```

```python
#create distplots
fig, ax = plt.subplots(ncols=6, nrows=5, figsize=(20,20))
index=0
ax = ax.flatten()

for col in df_temp.columns:
    sns.distplot(df[col],ax=ax[index])
    index+=1
plt.tight_layout(pad=0.5, w_pad=0.7, h_pad=5.0)
```

```python
# create boxplot
fig, ax = plt.subplots(ncols=6, nrows=5, figsize=(20,20))
index=0
ax = ax.flatten()

for col in df_temp.columns:
    sns.boxplot(y=col, data=df,ax=ax[index])
    index+=1
plt.tight_layout(pad=0.5, w_pad=0.7, h_pad=5.0)
```
